```python
#Lista 3 - Bruna Zacharias -- 2110485


import numpy as np
import pandas as pd

breast_cancer = pd.read_csv('/content/drive/MyDrive/class_breast_cancer.csv')
breast_cancer =pd.DataFrame(breast_cancer)
display(breast_cancer)


#Inspecione as colunas do conjunto de dados (Nome das colunas e tipo de dados)

breast_cancer.info()

#Escolha uma coluna com valores numéricos e apague todas as linhas que possuam valor nulo

breast_cancer['compacity'] = breast_cancer['compacity'].replace(0,np.nan)
breast_cancer['compacity'] = breast_cancer['compacity'].dropna()

#Calcule a média desta mesma coluna

media = breast_cancer['compacity'].mean()
print(media)

#Escolha um critério para eliminar as linhas que possuam valores muito distantes da média e crie um novo Dataframe.

eliminar = breast_cancer['compacity'] >= 0.02
eliminar2 = breast_cancer['compacity'] <= 0.7

breast_cancer = breast_cancer[eliminar]
breast_cancer = breast_cancer[eliminar2]

#Crie um novo Dataframe que contenha apenas 5 colunas do conjunto anterior.

otherbreast_cancer = breast_cancer.drop(['smoothness','radius','texture','area','perimeter'],axis = 1)

#Escolha duas colunas numéricas e crie uma nova coluna que siga a regra: (Coluna A - Coluna B)*número aleatório (entre 0 e 1).

otherbreast_cancer['novacoluna'] = otherbreast_cancer['fractal']-otherbreast_cancer['concavity']

otherbreast_cancer['novacoluna'] = otherbreast_cancer['novacoluna']*0.3

#Embaralhe aleatoriamente o dataframe (índices e valores).

breast_cancer_random = otherbreast_cancer.sample(frac=1)

#Ordene o Dataframe seguindo um critério definido por você (Explique o critério)


otherbreast_cancer = otherbreast_cancer[['novacoluna','concavity','fractal','symmetry','target']] #organizei as colunas na ordem desejada, sendo a nova coluna a primeira, seguida por concavity, fractal, simmetry e por fim target.
otherbreast_cancer.sort_values(by = ['concavity']).reset_index(drop=True) #Após mudar a ordem das colunas, organizei a simetria em ordem crescente, por fim renumerando os index para acompanhar ordem crescente da simetria


#Escolha uma coluna numérica e um coluna com strings e utilize a função Groupby para mostrar os totais da coluna.

grupo = otherbreast_cancer.groupby(['concavity','target']).count()

#Utilize a função MAPS para fazer algum tipo de substituição no Dataframe (explique)
otherbreast_cancer['target'] = otherbreast_cancer['target'].map({'M':'Metastasia','B':'Luminal'}) #coloquei no lugar de M, coloquei metastasis e no lugar de B coloquei luminal

#Exporte o último Dataframe para um arquivo CSV 

otherbreast_cancer.to_csv('cancerdemama.csv')

#Exporte o último Dataframe para um arquivo JSON 
otherbreast_cancer.to_json('cancerdemama.json')




```
